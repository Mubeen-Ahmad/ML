# Normalization

### Normalization is a scaling Technique in which the values are rescaled between  the range ( 0 to 1 )

# Most Common Types of Normalization
* ##  Min Max 
* ##  Mean
* ##  Max Absolute
* ##  Robust 

```python
import pandas as pd
from sklearn.preprocessing import Normalizer, MinMaxScaler ,MaxAbsScaler ,RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
import numpy as np
```

```python
df = load_wine()
df = pd.DataFrame(df.data,columns=df.feature_names)
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)
xtrain
```

# MinMax

## formulae 

* ## X = x(i) - x_min / x_max - x_min

```python
Mimx = MinMaxScaler()
Mimx.fit(xtrain)
Mimx_xtrain =  Mimx.transform(xtrain)
Mimx_xtest = Mimx.transform(xtest)
Mimx_xtrain = pd.DataFrame(Mimx_xtrain,columns=xtrain.columns)
Mimx_xtrain
```

# Means

## formula
* ## X = x(i) -  xmean / xmax - xmin

### Note Sklearn don't have Means Function  

```python
col = xtrain.columns

m = dict()

for i in col:
    m.update({i:  {"mean":np.round(np.mean(xtrain[i]),2)}  })
    m[i].update({"max":max(xtrain[i])})
    m[i].update({"min":min(xtrain[i])})


val = dict()
for i in col:
    val.update({i: {} })

no = 0
for c in col:
    for i in xtrain[c]:
        val[c].update({no:(i-m.get(c).get("mean"))/(m.get(c).get("max")-m.get(c).get("min"))})
        no += 1

lst = len(col)
for i in range(lst):
    st = (f'{col[i]}=[j for j in val.get(col[i]).values()]')
    exec(st)

    
for i in range(len(col)):
     print(i,)
```
