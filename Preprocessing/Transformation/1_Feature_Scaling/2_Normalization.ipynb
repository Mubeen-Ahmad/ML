# Normalization

### Normalization is a scaling Technique in which the values are rescaled between  the range ( 0 to 1 )

# Most Common Types of Normalization
* ##  Min Max 
* ##  Mean
* ##  Max Absolute
* ##  Robust 

```python
import pandas as pd
from sklearn.preprocessing import Normalizer, MinMaxScaler ,MaxAbsScaler ,RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
import numpy as np
```

```python
df = load_wine()
df = pd.DataFrame(df.data,columns=df.feature_names)
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)
xtrain
```

# MinMax

### Data are scaled in the range 0 to 1


## formulae 

* ## X = x(i) - x_min / x_max - x_min

```python
Mimx = MinMaxScaler()
Mimx.fit(xtrain)

Mimx_xtrain =  Mimx.transform(xtrain)
Mimx_xtest = Mimx.transform(xtest)

Mimx_xtrain = pd.DataFrame(Mimx_xtrain,columns=xtrain.columns)
Mimx_xtrain
```

```python
xtrain.describe()
```

```python
Mimx_xtrain.describe()
```

# Means

### Data are scaled in the range -1 to 1
* ### but we are used Standarization instead of Means 

## formula
* ## X = x(i) -  xmean / xmax - xmin

### Note Sklearn don't have Means Function  

```python

col = xtrain.columns

m = dict()

for i in col:
    m.update({i:  {"mean":np.round(np.mean(xtrain[i]),2)}  })
    m[i].update({"max":max(xtrain[i])})
    m[i].update({"min":min(xtrain[i])})

print(m)
```

```python
val = dict()
for i in col:
    val.update({i: {} })

print(val)
```

```python
no = 0
for c in col:
    for i in xtrain[c]:
        val[c].update({no:(i-m.get(c).get("mean"))/(m.get(c).get("max")-m.get(c).get("min"))})
        no += 1
print(val)
```

```python
al = val.get(col[0])
ml = val.get(col[1])
new = pd.DataFrame(al.values(),columns=["alcohol"])
new["malic_acid"] = ml.values()
new
```

```python
np.round(new.describe(),1)
```

# Max Absolute

### when maximum 0 in the data then we are used Max Absolute

## formulae

* ## X = x(i) / |x_max|


```python
from sklearn.preprocessing import MaxAbsScaler
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)

ma = MaxAbsScaler()
ma.fit(xtrain)
```

```python
ma_xtrain = ma.transform(xtrain)
ma.transform(xtest)

ma_xtrain= pd.DataFrame(ma_xtrain,columns= xtrain.columns)
ma_xtrain
```

```python
np.round(ma_xtrain.describe(),1)
```

# Robust

### Handling the outliers

## formulae

* ## X = x(i) - x_median / x75 percentile - x25 percentile

```python
from sklearn.preprocessing import RobustScaler
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)

rc = RobustScaler()
rc.fit(xtrain)

new = rc.transform(xtrain)
rc.transform(xtest)

new = pd.DataFrame(new)
new
```

```python
np.round(new.describe(),1)
```
