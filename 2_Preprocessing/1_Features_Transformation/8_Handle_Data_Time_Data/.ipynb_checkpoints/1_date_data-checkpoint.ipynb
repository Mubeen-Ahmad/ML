# Change data object

# Extract year 

## dt.year

```python
date["date_year"] = date["date"].dt.year
```

```python
date.head()
```

# Extract Month
## dt.month

```python
date["date_month_no"] = date["date"].dt.month
```

```python
date.head()
```

# Extract Month with name

## dt.month_name()

```python
date["date_month_name"] = date["date"].dt.month_name()
```

```python
date.head()
```

# Extract Days
## dt.day

```python
date["date_day"] = date["date"].dt.day
```

```python
date.head()
```

# day of week
## dt.daypfweek

```python
date["date_day_of_weak"] = date["date"].dt.dayofweek
date.head()
```

# day in name
## dt.date_name()

```python
date["date_day_name"] = date["date"].dt.day_name()
date.head()
```

## check specfic value 

```python
date["date_is_weekend"] = np.where(date.date_day_name.isin(["Sunday","Saturday"]),1,0)
```

```python
date.head()
```

# Extract week of the year
## dt.week / dt.isocalendar().week


```python
date["date_week"] = date["date"].dt.isocalendar().week
date.head()
```

# Extract Quarater

```python
date["quarter"] = date["date"].dt.quarter
date.head()
```

# Extract Semester

## 1 year = 2 Semester

```python
date["semester"] = np.where(date.quarter.isin([1,2]),1,2)
date.head()
```
