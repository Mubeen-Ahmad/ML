# A Box Cox is a transformation of non-normal dependent variables into a normal shape.

# Box-Cos are convert from NOn-Distribution Data to Distribution Data 

* # The exponent here is a variable called lambda. that varies over the range of `-5 to 5`
* # and in the process of searching, we exmine all value of lambda.
* # Finally we choose the optimal value (resulting in the best approximation to a normal distribution for your variable)

# Feature Scaling
## Apply Normalization --> mean scale

```python
def mean(dt):
    col = dt.columns
    m = dict()

    for i in col:
        m.update({i:  {"mean":np.round(np.mean(dt[i]),2)}  })
        m[i].update({"max":max(dt[i])})
        m[i].update({"min":min(dt[i])})
        
    val = dict()
    for i in col:
        val.update({i: {} })
    
    no = 0
    for c in col:
        for i in dt[c]:
            val[c].update({no:(i-m.get(c).get("mean"))/(m.get(c).get("max")-m.get(c).get("min"))})
            no += 1

    ds = {}
    for k,i in val.items():
        ds.update({k:list(i.values())})
    
    return pd.DataFrame(ds)
```

```python
X_TR =  mean(X_train)
X_TS =  mean(X_test)
```

```python
for i in X_TR.columns:
    
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.distplot(X_TR[i])
    plt.title(i)
    
    plt.subplot(122)
    stats.probplot(X_TR[i],dist="norm",plot=plt)
    plt.title(i)
    plt.show()
```
