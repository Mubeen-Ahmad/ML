# reciprocal Transformer
## In this transformation, x will replace by the inverse of x (1/x).

## The reciprocal transformation will give little effect on the shape of the distribution. This transformation can be only used for non-zero values.



```python
import pandas as pd 
import numpy as np 
import scipy.stats as stats
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
```

```python
data = {
    "no":[5000,4000,100,400,300]
}

df = pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(df["no"])
plt.show()
```

```python
tr = FunctionTransformer(func=lambda x:1/x)
tr_data = tr.fit_transform(df["no"])
tr_data
```

```python
plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(tr_data)
plt.show()
```

# Square Transform

## Square Transform are good work on Left Skewed data

```python
data = {
    "no":[5000,4000,100,400,300]
}

df = pd.DataFrame(data)

plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(df["no"])
plt.show()
```

```python
tr = FunctionTransformer(func=lambda x:x**2)
tr_data = tr.fit_transform(df["no"])

plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(tr_data)
plt.show()
```

# Arcsine Transformation
## An arcsine transformation can be used to “stretch out” data points that range between the values 0 and 1.

```python
data = {
    "no":[5000,4000,100,400,300]
}
np.arcsin([0,1,2,3])

# only working on 0,1 values
```

```python
data = {
    "no":[0.00855,0.6653,1.8979,1.2231200,1.300]
}

df = pd.DataFrame(data)

plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(df["no"])
plt.show()

tr_data = np.arcsin(df["no"])

plt.figure(figsize=(14,5))
plt.subplot(121)
sns.distplot(tr_data)
plt.show()
```
