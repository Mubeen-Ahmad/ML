# Normalization

### Normalization is a scaling Technique in which the values are rescaled between  the range ( 0 to 1 )

# Most Common Types of Normalization
* ##  Min Max 
* ##  Mean
* ##  Max Absolute
* ##  Robust 
* ##  Normalizer
* ##  QuantileTransformer
* ##  Unit Vector Scaler

```python
import pandas as pd
from sklearn.preprocessing import Normalizer, MinMaxScaler ,MaxAbsScaler ,RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
import numpy as np
```

```python
df = load_wine()
df = pd.DataFrame(df.data,columns=df.feature_names)
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)
xtrain
```

# MinMax

### Data are scaled in the range 0 to 1


## formulae 

* ## X = x(i) - x_min / x_max - x_min

```python
Mimx = MinMaxScaler()
Mimx.fit(xtrain)

Mimx_xtrain =  Mimx.transform(xtrain)
Mimx_xtest = Mimx.transform(xtest)

Mimx_xtrain = pd.DataFrame(Mimx_xtrain,columns=xtrain.columns)
Mimx_xtrain
```

# Means

### Data are scaled in the range -1 to 1
* ### but we are used Standarization instead of Means 

## formula
* ## X = x(i) -  xmean / xmax - xmin

### Note Sklearn don't have Means Function  
## So Created Manual Function

```python
class Mean:
    
    columns_means = {}    
    min_val = {}
    max_val = {}
    
    new_data = None

    
    
    def __init__(self,data):
        self.data = data
        self.get_columns()
        self.get_min_max()
        self.Apply()
        
    def get_columns(self):
        
        for i in self.data.columns:
            Mean.columns_means.update({i:self.data[i].mean()})
    
    def get_min_max(self):
        
        for i in self.columns_means.keys():
            Mean.min_val.update({i:self.data[i].min()})
        
        for i in self.columns_means.keys():
            Mean.max_val.update({i:self.data[i].max()})
    
    def Apply(self):
        new = []
        for k,v in self.columns_means.items():
            
            col = k
            mean = v
            min_ = self.min_val.get(k)
            max_ = self.max_val.get(k)
            new.append(np.array((self.data[col]-mean) / (max_- min_)))
        
        new_data = {}
        for i in range(len(self.columns_means)):
            new_data.update({list(self.columns_means.keys())[i] : new[i]})
        
        
        new_data = pd.DataFrame(new_data)
        Mean.new_data = new_data
```

```python
Mean(x).new_data
```

```python
np.round(Mean(x).new_data.describe(),1)
```

# Max Absolute

### when maximum 0 in the data then we are used Max Absolute

## formulae

* ## X = x(i) / |x_max|


```python
from sklearn.preprocessing import MaxAbsScaler
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)

ma = MaxAbsScaler()
ma.fit(xtrain)
```

```python
ma_xtrain = ma.transform(xtrain)
ma.transform(xtest)

ma_xtrain= pd.DataFrame(ma_xtrain,columns= xtrain.columns)
ma_xtrain
```

```python
np.round(ma_xtrain.describe(),1)
```

# Robust

### Handling the outliers

## formulae

* ## X = x(i) - x_median / x75 percentile - x25 percentile

```python
from sklearn.preprocessing import RobustScaler
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)

rc = RobustScaler()
rc.fit(xtrain)

new = rc.transform(xtrain)
rc.transform(xtest)

new = pd.DataFrame(new)
new
```

```python
np.round(new.describe(),1)
```

# QuantileTransformer


```python
from sklearn.preprocessing import QuantileTransformer
xtrain,xtest,ytrain ,ytest = train_test_split(x,y,test_size=20,random_state=10)
```

```python
QT=QuantileTransformer(n_quantiles=20)
```

```python
QT.fit(xtrain)
```

```python
QT_data = pd.DataFrame(QT.fit_transform(xtest))
QT_data
```
