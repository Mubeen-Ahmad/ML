# System of Linear Equations

<h2><p>A system of linear equations is two or more linear equations that are being solved simultaneously. </p></h2>

<h3> There are three possible outcomes that you may encounter when working with these systems: </h3>

<ol>
<li>one solution</li>
<li>infinite solutions</li>
<li>no solution</li>
</ol>

<h2>One Soloution</h2>
<p>If the system in two variables has one solution, it is an ordered pair that is a solution to BOTH equations. <br>
In other words, when you plug in the values of the ordered pair it makes BOTH equations TRUE</p>

# Example

# $$ 6x - 3 = 21 $$
# let x = 4
# $$ 6(4) - 3 = 21 $$
# $$ 24 - 3 = 21 $$
# $$ 21 = 21 $$

 

<h2>Infinite solutions</h2>
<p>If a linear equation has the same variable term and the same constant value on both sides of the equation, it has infinitely many solutions.</p>

# Example

# $$ x + y = 2 $$
# $$ x = 2 - y $$
# soloution 1
# let y = 3
# $$ x = 2 - 3 $$
# $$ x = -1 $$
# $$ x + y = 2 $$
# $$ -1 + 3 = 2 $$
# $$ 2 = 2 5$$ 

# soloution 2
# let y = 5
# $$ x + y = 2 $$
# $$ x + 5 = 2 $$
# $$ x = 2 - 5 $$
# $$ x = -3 $$
# $$ x + y = 2 $$
# $$ -3 + 5 = 2 $$
# $$ 2 = 2 $$


<h2>No Soloution</h2>
<p> If the two lines are parallel to each other, they will never intersect.  This means they do not have any points in common.  In this situation, you would have no solution. </p>

# Example

# $$ 3x + 8 = 1 +3x $$
# let x = 3
# $$ 3(3) + 8 = 1 +3(3) $$
# $$ 9 + 8 = 1 + 9 $$
# $$ 9 + 8 = 1 + 9 $$

# $$ 17 = 10 $$
# $$ No \,Soloution $$

<h1>Consistent system</h1>
<p> <h2>A consistent system is a system that has at least one solution</h2> </p> 
<br>
<h3> Example One solution , Infinite solutions </h3>

<h1>Inconsistent System</h1>
<p> <h2>An inconsistent system is a system that has no solution. </h2> </p>

<h1> Solve by the Cross Multiplication Method </h1><br>
<h1>$$ 2x +y = 5 $$</h1>

<h1>$$ 2x +y -5 = 0 $$</h1>
